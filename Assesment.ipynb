```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
```

```python
param={
    'svm':{
        'model':SVC(gamma='auto'),
        'params':{
            'C':[1,10,20],
            'kernel':['rbf','linear']
        }
    },
    
    'logistic_regression':{
        'model':LogisticRegression(),
        'params':{
            'C':[1,5,10]
        }
        
    },
    
    'random_forest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[5,10,20,40,50]

        }
    },
    'gaussianNB':{
        'model':GaussianNB(),
        'params':{
            
        }
    },
    'multinomialNB':
        {
            'model':MultinomialNB(),
            'params':{}
            
        }
    }
    
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
s=[]
for m_name,mp in param.items():
    clf=GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf.fit(x_train,y_train)
    s.append({
        'model':m_name ,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
      
```

```python
pd.DataFrame(s)
```

```python
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
x_train_a,x_test_a,y_train_a,y_test_a=train_test_split(x,y,test_size=0.2)
```

```python
model=keras.models.Sequential([
    keras.layers.Dense(256,input_shape=(41,),activation='relu'),
    keras.layers.Dense(128,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(32,activation='relu'),
    keras.layers.Dense(3,activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
```

```python
model.fit(x_train_a,y_train_a,epochs=50)
```

```python
model.evaluate(x_test,y_test)
```

### Ans: In this model after analysing the data and doing data preprocessing, I chose to run all the training data through GridSearchCv which is a quick and efficient module to judge the best ML model required to get more accurate model and as the results shows RandomForestClassifier showed the best outcome of all the tested ML models.

## Q2. What are the different tuning methods used for the algorithm?
### Ans: 
### 1.Firstly I checked for the NA values that were present in the dataset, and the NA to total rows in dataset ratio was very small so dropping the NA values directly will not impact the model in any negative manner.
### 2.I dropped the columns which will not impact the model predictions like kepler names etc, so it does not affect the output by including outliers.
### 3.Then using correlation, the data columns were discovered that had the most impact on the test variable,though dropping the columns will less correlation resulted in poor accuracy as there was a significant data loss when combined together. 
### 4.Next scaling the data also decreased the accuracy so I chose to use unscaled data and it can also be noticed that the values are not very large to be scaled and accuracy improved significantly.
